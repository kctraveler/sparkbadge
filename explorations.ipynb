# Build status: color mesh

```python
success, pending, failure = 0.7, 0.4, 0.1
fig = plt.figure(figsize=[9, 3], frameon=False)
plt.pcolormesh([[success, success, failure, success, failure, pending, success, failure, failure, pending, failure, success]],vmin=0, vmax=1, cmap='Spectral')
plt.axis('off')
plt.savefig('bars.svg', bbox_inches='tight', pad_inches=0)
```

# Quantitative data: sparkline

```python
x = [1, 2, 3, 4, 5]
y = [1, 3.3, 4, 6, 6.6]
fig = plt.figure(figsize=[9, 3], frameon=False)
plt.axis('off')
plt.plot(x,y, linewidth=5)
plt.arrow(5, 6.6, 0.5, 0.3, length_includes_head=True, head_width=0.5, edgecolor=None)
plt.savefig('plot.svg', bbox_inches='tight', pad_inches=0)
```

# Drawing the badge label

```python
fig = plt.figure(figsize=[9, 3], frameon=False)
plt.axis('off')
plt.text(0, 0, "wasup", color='white', backgroundcolor='purple', fontsize=150, ma='center')
plt.savefig('text.svg', bbox_inches='tight', pad_inches=0)
```
